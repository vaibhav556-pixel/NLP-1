```python
## Question 6: Extract Emails using Regex
import re
text='Contact support@xyz.com and hr@xyz.com'
print(re.findall(r'\S+@\S+', text))
```

```python
## Question 7: Tokenization & Frequency Distribution
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('punkt')
text='NLP is interesting'
tokens=word_tokenize(text)
print(tokens)
```

```python
## Question 8: Proper Noun Annotator
import nltk
from nltk import pos_tag, word_tokenize
nltk.download('averaged_perceptron_tagger')
text='John works at Google'
tokens=word_tokenize(text)
print([w for w,t in pos_tag(tokens) if t=='NNP'])
```

```python
## Question 9: Word2Vec using Gensim
from gensim.models import Word2Vec
sentences=[['nlp','is','fun'],['word2vec','model']]
model=Word2Vec(sentences, min_count=1)
print(model.wv['nlp'])
```
